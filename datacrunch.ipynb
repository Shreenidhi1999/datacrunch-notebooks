```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
yt1
```

```python
yt1.shape
```

```python
model=Sequential()
model.add(Dense(200,activation='relu',kernel_initializer='he_normal',input_dim=29))
```

```python
model.add(Dense(200,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(100,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(100,activation='relu',kernel_initializer='he_normal'))
model.add(Dropout(0.4))
model.add(Dense(30,activation='relu',kernel_initializer='he_normal'))
```

```python
model.compile(optimizer='adam',metrics=['accuracy'],loss='mean_squared_error')
model.summary()
```

```python
model.fit(xt,yt1,epochs=30)
```
