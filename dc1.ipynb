```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("X_train.csv")
test_data=pd.read_csv("X_test.csv")
data1=pd.read_csv("y_train.csv")
```

```python
data.head()
```

```python
test_data.head()
```

```python
xtrain=data.drop('id',axis=1)
xtrain=xtrain.drop('moons',axis=1)
```

```python
xtrain
```

```python
xtest=test_data.drop('id',axis=1)
xtest=xtest.drop('moons',axis=1)
```

```python
xtest
```

```python
from sklearn.decomposition import PCA
pc=PCA(n_components=2)
xtrain=pc.fit_transform(xtrain)
xtest=pc.fit_transform(xtest)
print(xtrain)
```

```python
xtest
```

```python
from sklearn.model_selection import train_test_split
xt,xte,yt,yte=train_test_split(xtrain,data1,test_size=0.2,random_state=7)
```

```python
xt.shape
```

```python
xte.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xt,yt)
ypred=lr.predict(xte)
```

```python
from sklearn.metrics import mean_squared_error as mse
np.sqrt(mse(ypred,yte))
```

```python
yfinal=lr.predict(xtest)
```

```python
yfinal
```

```python
yfinal.max()
```

```python
yfinal=pd.DataFrame(yfinal,columns=['target_r','target_g','target_b'])
```

```python
yfinal
```
